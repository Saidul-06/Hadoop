```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset.csv')
data
```

```python
data = data.drop(labels="car name", axis=1)
```

```python
data = data.dropna(how = 'any')
data
```

```python
data.isnull().sum()
```

```python
x = data.drop(labels="origin", axis=1)
x
```

```python
y = data.origin
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['origin']=le.fit_transform(data['origin'])
data
```

```python
from sklearn.model_selection import train_test_split
xt,xtt,yt,ytt = train_test_split(x,y,test_size=0.5,stratify=y,random_state=1)
```

```python
x = data.iloc[:,:-1]
x
```

```python
y = data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
xt,xtt,yt,ytt = train_test_split(x,y,test_size=0.5,stratify=y,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(xt,yt)
```

```python
from sklearn import metrics
y_pred=log.predict(xtt)
metrics.accuracy_score(ytt,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=1000, random_state=0)
forest.fit(xt,yt)
```

```python
from sklearn import metrics
y_pred=forest.predict(xtt)
metrics.accuracy_score(ytt,y_pred)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(xt,yt)
```

```python
from sklearn import metrics
y_pred =svm.predict(xtt)
metrics.accuracy_score(ytt, y_pred)
```

```python
from sklearn import tree
X = xt
Y = yt
tree = tree.DecisionTreeClassifier()
tree = tree.fit(X,Y)
```

```python
from sklearn import metrics
y_pred=tree.predict(xtt)
metrics.accuracy_score(ytt,y_pred)
```
