```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset.csv')
data
```

```python
data = data.drop(labels="FileName", axis=1)
data = data.drop(labels="Date", axis=1)
data = data.drop(labels="SegFile", axis=1)
data
```

```python
data = data.dropna(how = 'any')
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['NSP']=le.fit_transform(data['NSP'])
data
```

```python
x = data.iloc[:,:-1]
x
```

```python
#2=1, otherwise=0

y = data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
xt,xtt,yt,ytt = train_test_split(x,y,test_size=0.5,stratify=y,random_state=1)
```

***Logistic = 89%***

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(xt,yt)
```

```python
from sklearn import metrics
y_pred=log.predict(xtt)
metrics.accuracy_score(ytt,y_pred)
```

***Random Forest***

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=1000, random_state=0)
forest.fit(xt,yt)
```

```python
from sklearn import metrics
y_pred=forest.predict(xtt)
metrics.accuracy_score(ytt,y_pred)
```
